```python
csv = csv[csv["Street Name"].isin(["MASS AVE", "MASSACHUSETTS AVE", "MASSACHUSETTS AVENUE"]) | csv["Cross Street"].isin(["MASS AVE", "MASSACHUSETTS AVE", "MASSACHUSETTS AVENUE"])]
```

```python
from datetime import datetime
councilterm = csv#[csv["Date Time"] >= datetime(2018, 1, 1)]
councilterm["Date Time"]
```

```python
sum(councilterm["V2 Model"] == "BIKE")
```

```python
councilterm = councilterm.fillna({"P2 Non Motorist Desc": "", "P1 Non Motorist Desc": "", "May involve cyclist": 0, "Object 1": "", "P1 Injury": "", "P2 Injury": ""})
cyclists = councilterm[(councilterm["May involve cyclist"] > 0) & (councilterm["Object 1"] != "") & (
    (councilterm["P2 Non Motorist Desc"] == "CYCLIST") | (councilterm["P1 Non Motorist Desc"] == "CYCLIST"))]
cyclists
```

```python
summary = cyclists[["Date Time", "Object 1", "Object 2", "Street Number", "Street Name", "Cross Street", "P1 Injury", "P1 Non Motorist Desc", "P2 Injury", "P2 Non Motorist Desc", "P1 Sex", "Street or Intersection"]]
#summary = summary[summary["Street or Intersection"] != "INTERSECTION"]

print(len(list(summary.iterrows())))
with pandas.option_context('display.max_rows', None, 'display.max_columns', None):
    display(summary)
```

```python
from IPython.display import display, Markdown, Latex

i = 0
result = ""
for _, (dt, vehicle, _, street_number, street, street2, injury1, _, injury2, _, sex, intersection) in reversed(sorted(list(summary.iterrows()), key=lambda l: l[1][0])):
    i += 1
    date = dt.date()
    injury = injury1 if injury1 != "NO INJURY" else injury2
    if vehicle.startswith("LIGHT TRUCK"):
        vehicle = "LIGHT TRUCK"
    #if injury in ("NON FATAL INJURY - POSSIBLE", "UNKNOWN", ""):
    #    continue
    if injury == "UNKNOWN":
        injury = "UNKNOWN IF INJURED"
    injury = ", " + injury
    preventable = ", potentially preventable by protected bicycle lanes" if intersection == "NOT INTERSECTION" else ""
    result += f"{i}. On {date} at {street_number} {street}/{street2} the driver of a {vehicle} crashed into a bicycle rider{preventable}.\n"
display(Markdown(result))
```
